```python
import numpy as np
import msicpe.san as san
from plotly import express as px
import scipy.signal as signal

signal_dict = np.load("signal_4.npz")   


s = signal_dict["s"]               
t = signal_dict["t"]               


fig_time = px.line(
    x=t,
    y=s,
    labels={"x": "Temps (s)", "y": "Amplitude"},
    title="Signal en fonction du temps"
)



f, S = san.trans_fourier(s, t)


fig_spectre = px.line(
    x=f,
    y=np.abs(S),
    labels={"x": "Fr√©quence (Hz)", "y": "|S(f)|"},
    title="Spectre du signal"
)
log_S = 20 * np.log10(np.abs(S))
fig_spectre_log = px.line(
    x=f,
    y=log_S,
    labels={"x": "Fr√©quence (Hz)", "y": "Amplitude (dB)"},
    title="Spectre du signal (√©chelle logarithmique)"
)

fig_spectre_log.show()
fig_time.show()
fig_spectre.show()

#nombre d'harmonqiue pour detecter la voyelle , comparer les caract√©risqtique sspectrales , mesurer comment deux voyelles sont proches bas√©es sur les amplitudes des harmoniques
```
